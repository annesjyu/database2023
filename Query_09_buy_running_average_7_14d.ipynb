```
import pandas as pd

data = pd.read_csv('daily_prices_rolling_average.csv')
data
```

```
import matplotlib.pyplot as plt

# make sure trade_date is datetime
data['trade_date'] = pd.to_datetime(data['trade_date'])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')

# plot close price
ax.fill_between(data['trade_date'], data['close_price'], label="Close Price", color="gray", alpha=0.3)

# plot running 7-day average
ax.plot(data['trade_date'], data['running_avg_14d'], label="14-Day Running Avg", color='coral')

# format x-axis
ax.xaxis.set_major_locator(plt.MaxNLocator(8))  # show about 8 ticks
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# labels & legend
ax.set_xlabel("Trade Date")
ax.set_ylabel("Price")
ax.legend()
plt.title("Close Price vs 7-Day Running Average")

plt.show()
```
